```python
# --- Imports ---
import numpy as np
import matplotlib.pyplot as plt

# --- Parameters ---
n0 = 1.0
e = 1.0
eps0 = 1.0

A = 1.0                  # amplitude of n1
k = 1.0                  # wavenumber
omega_pos = 1.0          # omega for case omega/k > 0
omega_neg = -1.0         # omega for case omega/k < 0

# n1(x,t) = A * exp(i(kx - wt))
def n1(x, t, omega):
    return A * np.exp(1j*(k*x - omega*t))

# Derived quantities
def v1(x, t, omega):
    return (omega/k) * n1(x, t, omega) / n0

def E1(x, t, omega):
    return (1j * e * n1(x, t, omega)) / (k * eps0)

def phi_i(x, t, omega):
    return -(e * n1(x, t, omega)) / (k**2 * eps0)

# phase in degrees
def phase(z):
    return np.angle(z, deg=True)
```

```python
# --- (a) Sine-wave representation vs x at t = 0 ---

x_vals = np.linspace(0, 4*np.pi/k, 400)
t0 = 0

n1_real_x = np.real(n1(x_vals, t0, omega_pos))
v1_real_x = np.real(v1(x_vals, t0, omega_pos))
E1_real_x = np.real(E1(x_vals, t0, omega_pos))
phi_real_x = np.real(phi_i(x_vals, t0, omega_pos))

plt.figure(figsize=(10,6))
# plt.plot(x_vals, n1_real_x, label="Re[n1]")
plt.plot(x_vals, v1_real_x, label=r'$\text{Re[}v_1\text{]}$')
plt.plot(x_vals, E1_real_x, label=r'$\text{Re[E}_1\text{]}$')
plt.plot(x_vals, phi_real_x, label=r'$\text{Re[}\phi_1\text{]}$')
plt.title("(a) Sine-Wave Representation vs x at t = 0")
plt.xlabel("x")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()
```

```python
# --- (b) Sine-wave representation vs t at x = 0, omega/k > 0 ---

t_vals_pos = np.linspace(0, 4*np.pi/omega_pos, 400)
x0 = 0

n1_real_t_pos = np.real(n1(x0, t_vals_pos, omega_pos))
v1_real_t_pos = np.real(v1(x0, t_vals_pos, omega_pos))
E1_real_t_pos = np.real(E1(x0, t_vals_pos, omega_pos))
phi_real_t_pos = np.real(phi_i(x0, t_vals_pos, omega_pos))

plt.figure(figsize=(10,6))
# plt.plot(t_vals_pos, n1_real_t_pos, label="Re[n1]")
plt.plot(t_vals_pos, v1_real_t_pos, label=r'$\text{Re[}v_1\text{]}$')
plt.plot(t_vals_pos, E1_real_t_pos, label=r'$\text{Re[E}_1\text{]}$')
plt.plot(t_vals_pos, phi_real_t_pos, label=r'$\text{Re[}\phi_1\text{]}$')
plt.title("(b) Sine-Wave Representation vs t at x = 0 (/k > 0)")
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()
```

```python
# --- (c) Sine-wave representation vs t at x = 0, omega/k < 0 ---

t_vals_neg = np.linspace(0, 4*np.pi/abs(omega_neg), 400)

n1_real_t_neg = np.real(n1(x0, t_vals_neg, omega_neg))
v1_real_t_neg = np.real(v1(x0, t_vals_neg, omega_neg))
E1_real_t_neg = np.real(E1(x0, t_vals_neg, omega_neg))
phi_real_t_neg = np.real(phi_i(x0, t_vals_neg, omega_neg))

plt.figure(figsize=(10,6))
# plt.plot(t_vals_neg, n1_real_t_neg, label="Re[n1]")
plt.plot(t_vals_neg, v1_real_t_neg, label=r'$\text{Re[}v_1\text{]}$')
plt.plot(t_vals_neg, E1_real_t_neg, label=r'$\text{Re[E}_1\text{]}$')
plt.plot(t_vals_neg, phi_real_t_neg, ':', linewidth=3, label=r'$\text{Re[}\phi_1\text{]}$')
plt.title("(c) Sine-Wave Representation vs t at x = 0 (/k < 0)")
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()
```

```python
# --- (a) Phase vs x at t = 0 ---
x_vals = np.linspace(0, 4*np.pi/k, 400)
t0 = 0

ph_v1_x = phase(v1(x_vals, t0, omega_pos))
ph_E1_x = phase(E1(x_vals, t0, omega_pos))
ph_phi_x = phase(phi_i(x_vals, t0, omega_pos))

plt.figure(figsize=(10,6))
plt.plot(x_vals, ph_v1_x, label="v1")
plt.plot(x_vals, ph_E1_x, label="E1 (90 shifted)")
plt.plot(x_vals, ph_phi_x, label=r'$\phi_1$')
plt.title("Relative Phases vs x at t = 0")
plt.xlabel("x")
plt.ylabel("Phase (deg)")
plt.legend()
plt.grid()
plt.show()
```

```python
# --- (b) Phase vs t at x = 0, omega/k > 0 ---
t_vals = np.linspace(0, 4*np.pi/omega_pos, 400)
x0 = 0

ph_v1_t_pos = phase(v1(x0, t_vals, omega_pos))
ph_E1_t_pos = phase(E1(x0, t_vals, omega_pos))
ph_phi_t_pos = phase(phi_i(x0, t_vals, omega_pos))

plt.figure(figsize=(10,6))
plt.plot(t_vals, ph_v1_t_pos, label="v1")
plt.plot(t_vals, ph_E1_t_pos, label="E1 (90 shifted)")
plt.plot(t_vals, ph_phi_t_pos, label=r'$\phi_1$')
plt.title("Relative Phases vs t at x = 0  (omega/k > 0)")
plt.xlabel("t")
plt.ylabel("Phase (deg)")
plt.legend()
plt.grid()
plt.show()
```

```python
# --- (c) Phase vs t at x = 0, omega/k < 0 ---
t_vals = np.linspace(0, 4*np.pi/abs(omega_neg), 400)

ph_v1_t_neg = phase(v1(x0, t_vals, omega_neg))
ph_E1_t_neg = phase(E1(x0, t_vals, omega_neg))
ph_phi_t_neg = phase(phi_i(x0, t_vals, omega_neg))

plt.figure(figsize=(10,6))
plt.plot(t_vals, ph_v1_t_neg, label="v1")
plt.plot(t_vals, ph_E1_t_neg, label="E1 (90 shifted)")
plt.plot(t_vals, ph_phi_t_neg, label=r'$\phi_1$')
plt.title("Relative Phases vs t at x = 0  (omega/k < 0)")
plt.xlabel("t")
plt.ylabel("Phase (deg)")
plt.legend()
plt.grid()
plt.show()
```
